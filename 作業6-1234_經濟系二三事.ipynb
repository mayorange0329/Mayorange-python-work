```python
from matplotlib_venn import venn3
```

```python
v=venn3(subsets = (7, 8, 0, 0,1,0,0), set_labels = ('Group A', 'Group B', 'Group C'))
v.get_label_by_id('A').set_text("Social Science")
v.get_label_by_id('B').set_text("Commerce")
v.get_label_by_id('C').set_text("ECON")
plt.show()
```

作業4-字串型的互動金字塔

```python
from ipywidgets import interact

def love(x):
    for i in range(x):
        a = "I LOVE ECON " * (2*i - 1)
        print(f"{a:^100s}")
       
interact(love, x = (0,4))
print("                                            ヾ(●゜▽゜●)/")
```
