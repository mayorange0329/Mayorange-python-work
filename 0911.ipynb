```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
r1=10
r2=5
t=np.linspace(-2*np.pi,2*np.pi,200)
s=np.linspace(-2*np.pi,0,200)

x1=r1*np.cos(t)
y1=r1*np.sin(t)

x2=[-5,-5,-5]
y2=[0,2.5,4]

x3=[5,5,5]
y3=[0,2.5,4]

x5=[-1,0,1]
y5=[-1,0,-1]

x6=r2*np.cos(0.5*s)
y6=r2*np.sin(0.5*s)-2.5

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x5,y5)
plt.plot(x6,y6)
```
