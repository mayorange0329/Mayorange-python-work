## 1. Consumption Demand

給定兩期消費模型問題為

\begin{equation}
\begin{split}
\max \;\; & U  = u(C_t) + \beta u(C_{t+1}) \\
s.t. \;\; & C_t + \frac{C_{t+1}}{1+r_t} = Y_t + \frac{Y_{t+1}}{1+r_t}
\end{split}
\end{equation}

而效用函數為 $$u(C) = \sqrt{C},\; 其中 u'(C) = 0.5 C^{-0.5}$$

1. 請推導出 Euler Equation 為 $$ C_t^{-0.5} = \beta (1+r_t) C_{t+1}^{-0.5}$$

<table><tr><td bgcolor=#98C7DD>〈Solution 1〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

因消費只存在兩期，所以：
$$Optimal:u'(C_t) = \beta (1+r_t) u'(C_{t+1})$$ 
代入$u'(C) = 0.5\cdot C^{-0.5}$：
$$ C_t^{-0.5} = \beta (1+r_t) C_{t+1}^{-0.5}───Euler Equation$$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

2. 使用 Euler Equation 與限制式 $$C_t + \frac{C_{t+1}}{1+r_t} = Y_t + \frac{Y_{t+1}}{1+r_t}$$ 推導出 $$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ Y_t + \frac{Y_{t+1}}{1+r_t} \bigg]$$

<table><tr><td bgcolor=#98C7DD>〈Solution 2〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

將Euler Equation $ C_t^{-0.5} = \beta \cdot (1+r_t) \cdot C_{t+1}^{-0.5}$ 開$-0.5$ 次方根並整理：
$$ C_t = \beta^{-2} \cdot (1+r_t)^{-2} \cdot C_{t+1}$$
$$ C_{t+1} = C_t \cdot \beta^{2} \cdot (1+r_t)^2 $$
再代入限制式$C_t + \frac{C_{t+1}}{1+r_t} = Y_t + \frac{Y_{t+1}}{1+r_t}$：
$$C_t + C_t \cdot \beta^{2} \cdot (1+r_t) = Y_t + \frac{Y_{t+1}}{1+r_t}$$
整理後即可推導出$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ Y_t + \frac{Y_{t+1}}{1+r_t} \bigg]$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

3. 給定 $r_t = 0.05$, $Y_t = 0$, $Y_{t+1} = 105$, $\beta^2 (1+r_t) = 1$, 請推導出 


$$C_t = 50, S_t = -50$$

<table><tr><td bgcolor=#98C7DD>〈Solution 3〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

承上題結果$$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ Y_t + \frac{Y_{t+1}}{1+r_t} \bigg]$$
代入$r_t = 0.05$, $Y_t = 0$, $Y_{t+1} = 105$, $\beta^2 (1+r_t) = 1$：
$$C_t = \frac{1}{1+1} \cdot (0+\frac{105}{1+0.05})$$
得出
$$C_t = \frac{1}{2} \cdot 100 = 50$$
再將$C_t = 50$代入$S_t = Y_t - C_t$，得到：
$$S_t = Y_t - C_t= 0 - 50 = -50$$



<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

4. __[流動性限制]__ 給定現在存在流動性限制, 該消費者只能從銀行借入最多 20 單位，請推導出 $$C_t = 20, S_t = -20$$ 並解釋直覺

<table><tr><td bgcolor=#98C7DD>〈Solution 4〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

$$(直覺──由於第t期沒有Y，所以只能消費跟銀行最多能借到的20單位。)$$

$$ C_t + \frac{C_{t+1}}{1+r_t} = (1-\tau)Y_t + \frac{(1-\tau)Y_{t+1}}{1+r_t} $$

<table><tr><td bgcolor=#98C7DD>〈Solution 5〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

稅率$\tau$會讓可支配所得剩下 $ (1- \tau)\cdot Y$，使得限制式變成：

$$ C_t + \frac{C_{t+1}}{1+r_t} = (1-\tau)Y_t + \frac{(1-\tau)Y_{t+1}}{1+r_t} $$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

6. 接續 3, 請推導此時的 $C_t$ 為 

$$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ (1-\tau)Y_t + \frac{(1-\tau)Y_{t+1}}{1+r_t}  \bigg]$$

<table><tr><td bgcolor=#98C7DD>〈Solution 6〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

$$C_t + C_t \cdot \beta^{2} \cdot (1+r_t) = (1-\tau) Y_t + \frac{(1-\tau)Y_{t+1}}{1+r_t}$$
即可推導出此時的$C_t$ 為：
$$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ (1-\tau)Y_t + \frac{(1-\tau)Y_{t+1}}{1+r_t}  \bigg]$$

## 2. Equilibrium

給定兩期模型, 家戶 $i$ 的消費為 

$$C_t(i) = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ Y_t(i) + \frac{Y_{t+1}(i)}{1+r_t} \bigg]$$

現在假設有三個家戶, 亦即 $i = 1, 2, 3$。而他們相關的 Endowment 資訊如下 

- 家戶 1: $Y_t(1) = 30, Y_{t+1}(1) = 30$

- 家戶 2: $Y_t(2) = 50, Y_{t+1}(2) = 20$

- 家戶 3: $Y_t(3) = 20, Y_{t+1}(3) = 50$

請回答下列問題

1. 請推導總和消費 $C_t$ 如下

$$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 100 + \frac{ 100 }{1+r_t} \bigg] $$

<table><tr><td bgcolor=#98C7DD>〈Solution 1〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

$$C_t(1) = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 30 + \frac{30}{1+r_t} \bigg]$$
$$C_t(2) = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 50 + \frac{20}{1+r_t} \bigg]$$
$$C_t(3) = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 20 + \frac{50}{1+r_t} \bigg]$$

由於$C_t=C_t(1)+C_t(2)+C_t(3)$，因此推導出：
$$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 100 + \frac{ 100 }{1+r_t} \bigg] $$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


2. 請利用商品市場均衡條件 $Y_t = C_t$, 算出均衡利率為 

$$ r_t = \frac{1-\beta}{\beta}$$

<table><tr><td bgcolor=#98C7DD>〈Solution 2〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

將$Y_t = Y_t(1)+Y_t(2)+Y_t(3)=100=C_t$，代入$C_t = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 100 + \frac{ 100 }{1+r_t} \bigg] $，並整理：
$$100 = \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ 100 + \frac{ 100 }{1+r_t} \bigg] $$
$$ 1+\beta ^2\cdot (1+r_t) = 1+ \frac{1}{1+r_t}$$
$$(1+r_t)^2 = \frac{1}{\beta^2}$$
$$ r_t = \frac{1}{\beta} - 1$$
得出均衡利率為：
$$ r_t = \frac{1-\beta}{\beta}$$


<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>



3. 請推導家戶個別儲蓄函數為 

$$ S_t(i) = \frac{\beta^2\cdot(1+r_t)}{(1+\beta^2\cdot(1+r_t) )}  Y_t(i) -  \frac{1}{(1+\beta^2\cdot(1+r_t) )} \frac{Y_{t+1}(i)}{1+r_t}   $$

<table><tr><td bgcolor=#98C7DD>〈Solution 3〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

由於$ S_t(i) = Y_t(i) - C_t(i) $：
$$ S_t(i) = Y_t(i) - C_t(i) = Y_t(i) - \frac{1}{(1+\beta^2\cdot(1+r_t) )} \bigg[ Y_t(i) + \frac{Y_{t+1}(i)}{1+r_t} \bigg]$$
再進一步整理化簡：
$$ S_t(i) = (1 - \frac{1}{(1+\beta^2\cdot(1+r_t) )})\cdot Y_t(i) - \frac{1}{(1+\beta^2\cdot(1+r_t) )}\cdot \frac{Y_{t+1}(i)}{1+r_t} $$
最後得出：
$$ S_t(i) = \frac{\beta^2\cdot(1+r_t)}{(1+\beta^2\cdot(1+r_t) )}\cdot Y_t(i) -  \frac{1}{(1+\beta^2\cdot(1+r_t) )} \cdot \frac{Y_{t+1}(i)}{1+r_t}   $$


<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

4. 給定均衡利率下, 請驗證下面結果

$$ S_t(1) = 0, \; S_t(2) = \frac{\beta}{1+\beta} 30, \;  S_t(3) = -\frac{\beta}{1+\beta} 30   $$

<table><tr><td bgcolor=#98C7DD>〈Solution 4〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

將第2題得出的均衡利率 $ r_t = \frac{1-\beta}{\beta}$ ，代入第3題式子$ S_t(i) = \frac{\beta^2\cdot(1+r_t)}{(1+\beta^2\cdot(1+r_t) )}\cdot Y_t(i) -  \frac{1}{(1+\beta^2\cdot(1+r_t) )} \cdot \frac{Y_{t+1}(i)}{1+r_t}$ 並化簡，得出：
$$ S_t(i) = \frac{\beta}{1+\beta} \cdot Y_t(i) - \frac{\beta}{1+\beta}\cdot Y_{t+1}$$

$$ S_t(1) = \frac{\beta}{1+\beta} \cdot 30 - \frac{\beta}{1+\beta}\cdot 30= 0$$
$$ S_t(2) = \frac{\beta}{1+\beta} \cdot 50 - \frac{\beta}{1+\beta}\cdot 20= \frac{\beta}{1+\beta}\cdot30$$
$$ S_t(3) = \frac{\beta}{1+\beta} \cdot 20 - \frac{\beta}{1+\beta}\cdot 50=-\frac{\beta}{1+\beta}\cdot30$$

驗證 $$ S_t(1) = 0, S_t(2) = \frac{\beta}{1+\beta}\cdot30, S_t(3) = -\frac{\beta}{1+\beta}\cdot30   $$



<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


5. 接續 4, 請驗證總和儲蓄 $S_t = 0$

<table><tr><td bgcolor=#98C7DD>〈Solution 5〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

$S_t = S_t(1)+S_t(2)+S_t(3)$：

$$S_t = S_t(1)+S_t(2)+S_t(3)=0+\frac{\beta}{1+\beta}\cdot30-\frac{\beta}{1+\beta}\cdot30=0$$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

## 3. Multiple Periods

給定 4 期消費模型問題為

\begin{equation}
\begin{split}
\max \;\; & U  = u(C_t) + \beta u(C_{t+1}) + \beta^2 u(C_{t+2}) + \beta^3 u(C_{t+3}) \\
s.t. \;\; & Y_t = C_t + S_t \\ 
& Y_{t+1} + (1+r_t)S_t = C_{t+1} + S_{t+1} \\
& Y_{t+2} + (1+r_{t+1})S_{t+1} = C_{t+2} + S_{t+2} \\
& Y_{t+3} + (1+r_{t+2})S_{t+2} = C_{t+3} + S_{t+3} \\
\end{split}
\end{equation}

而效用函數為 $$u(C) = \ln C,\; 其中 u'(C) = 1/C$$

1. 請解釋為什麼 $S_{t+3} = 0$

<table><tr><td bgcolor=#98C7DD>〈Solution 1〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

因為消費共四期，第 $t+3$ 為最後一期，因此消費者不會在第 $t+3$ 期時存錢，$S_{t+3}=0$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


2. 給定 $r_{t} = r_{t+1} = r_{t+2} = r$, 請推導出終身預算限制式為 

$$ C_t + \frac{C_{t+1}}{1+r} + \frac{C_{t+2}}{(1+r)^2} + \frac{C_{t+3}}{(1+r)^3} =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3} $$

<table><tr><td bgcolor=#98C7DD>〈Solution 2〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


將限制式重新整理：


$$&Y_t = C_t + S_t$$
$$\frac{Y_{t+1}}{1+r} + S_t = \frac{C_{t+1}}{1+r} + \frac{S_{t+1}}{1+r}$$
$$\frac{Y_{t+2}}{(1+r)^2} + \frac{S_{t+1}}{1+r} = \frac{C_{t+2}}{(1+r)^2} +\frac{S_{t+2}}{(1+r)^2}$$
$$\frac{Y_{t+3}}{(1+r)^3} + \frac{S_{t+2}}{(1+r)^2} = \frac{C_{t+3}}{(1+r)^3}$$

把左式全部相加，右式也全部相加，並進行整理： 

$$Y_t+\frac{Y_{t+1}}{1+r} + S_t+\frac{Y_{t+2}}{(1+r)^2} + \frac{S_{t+1}}{1+r}+\frac{Y_{t+3}}{(1+r)^3} + \frac{S_{t+2}}{(1+r)^2}=C_t + S_t +\frac{C_{t+1}}{1+r} + \frac{S_{t+1}}{1+r}+\frac{C_{t+2}}{(1+r)^2} +\frac{S_{t+2}}{(1+r)^2}+\frac{C_{t+3}}{(1+r)^3}$$

即可推導出：

$$ C_t + \frac{C_{t+1}}{1+r} + \frac{C_{t+2}}{(1+r)^2} + \frac{C_{t+3}}{(1+r)^3} =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3} $$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


3. 給定 $r_{t} = r_{t+1} = r_{t+2} = r$,  請推導出 Euler Equations 為 

$$ C_t^{-1} = \beta (1+r) C_{t+1}^{-1}$$
$$ C_{t+1}^{-1} = \beta (1+r) C_{t+2}^{-1}$$
$$ C_{t+2}^{-1} = \beta (1+r) C_{t+3}^{-1}$$

<table><tr><td bgcolor=#98C7DD>〈Solution 3〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


最適解為：
$$u'(C_t) = \beta (1+r) u'(C_{t+1})$$
$$u'(C_{t+1}) = \beta (1+r) u'(C_{t+2})$$
$$u'(C_{t+2}) = \beta (1+r) u'(C_{t+3})$$

代入 $u'(C) = C^{-1}$ ，推導出 Euler equation 為：

$$ C_t^{-1} = \beta (1+r) C_{t+1}^{-1}$$
$$ C_{t+1}^{-1} = \beta (1+r) C_{t+2}^{-1}$$
$$ C_{t+2}^{-1} = \beta (1+r) C_{t+3}^{-1}$$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


4. 給定 $r_{t} = r_{t+1} = r_{t+2} = r$,  使用 Euler Equation 與限制式 $$ C_t + \frac{C_{t+1}}{1+r} + \frac{C_{t+2}}{(1+r)^2} + \frac{C_{t+2}}{(1+r)^3} =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+2}}{(1+r)^3} $$ 請推導出

$$ C_t = \frac{1}{1+\beta + \beta^2 + \beta^3}   X_t  ,\;\;  X_t =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+2}}{(1+r)^3}  $$

<table><tr><td bgcolor=#98C7DD>〈Solution 4〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


將上題求出的Euler Equation重新整理：
$$ C_{t+1}= \beta^1 (1+r)^1 C_t$$
$$ C_{t+2}= \beta^2 (1+r)^2 C_t$$
$$ C_{t+3}= \beta^3 (1+r)^3 C_t$$

將上述式子代入限制式$ C_t + \frac{C_{t+1}}{1+r} + \frac{C_{t+2}}{(1+r)^2} + \frac{C_{t+3}}{(1+r)^3} =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3} $並整理： 

$$ C_t + \frac{ \beta^1 (1+r)^1 C_t}{1+r} + \frac{\beta^2 (1+r)^2 C_t}{(1+r)^2} + \frac{\beta^3 (1+r)^3 C_t}{(1+r)^3} =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3} $$ 
$$ (1+\beta + \beta^2 + \beta^3) C_t = Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3} $$

最後推導出：
$$ C_t = \frac{1}{1+\beta + \beta^2 + \beta^3}   X_t  ,\;\;  X_t =  Y_t + \frac{Y_{t+1}}{1+r} + \frac{Y_{t+2}}{(1+r)^2} + \frac{Y_{t+3}}{(1+r)^3}  $$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>



5. 給定 $r_{t} = r_{t+1} = r_{t+2} = r$,  給定恆常所得為 $\overline{Y}$, 請算出 

$$ \overline{Y} = \frac{(1+r)^3}{(1+r)^3 + (1+r)^2 + (1+r) + 1} X_t $$

<table><tr><td bgcolor=#98C7DD>〈Solution 5〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


令$Y_t=Y_t+1=Y_t+2=Y_t+3=\overline{Y}$，並整理：

$$ \overline{Y}+\frac{\overline{Y}}{1+r}+\frac{\overline{Y}}{(1+r)^2}+\frac{\overline{Y}}{(1+r)^3}=X_t$$

$$ \overline{Y}(1+r)^3 +\overline{Y}(1+r)^2+\overline{Y}(1+r)+\overline{Y}=(1+r)^3 X_t$$

求出：
$$ \overline{Y} = \frac{(1+r)^3}{(1+r)^3 + (1+r)^2 + (1+r) + 1} X_t $$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


6. 假設 $1+r = 1/\beta$, 請算出 

$$ X_t = \frac{(1+r)^3 + (1+r)^2 + (1+r) + 1}{(1+r)^3}  \overline{Y} =  (1+\beta + \beta^2 + \beta^3) \overline{Y}$$

<table><tr><td bgcolor=#98C7DD>〈Solution 6〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

將第5題結果$ \overline{Y} = \frac{(1+r)^3}{(1+r)^3 + (1+r)^2 + (1+r) + 1} X_t $重新整理：

$$X_t=  \frac{(1+r)^3 + (1+r)^2 + (1+r) + 1}{(1+r)^3}  \overline{Y}  $$

$$X_t = (1 + (1+r)^{-1} + (1+r)^{-2} + (1+r)^{-3})\overline{Y} $$

再將 $1+r = \frac{1}{\beta} = \beta^{-1}$代入：

$$ X_t = \frac{(1+r)^3 + (1+r)^2 + (1+r) + 1}{(1+r)^3}  \overline{Y} =  (1+\beta + \beta^2 + \beta^3) \overline{Y}$$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>


7. 接續 6, 請算出 $$C_t = \overline{Y}$$ 並解釋經濟直覺

<table><tr><td bgcolor=#98C7DD>〈Solution 7〉-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>

將 $ X_t = (1+\beta + \beta^2 + \beta^3) \overline{Y}$ 代入第4題結果$ C_t = \frac{1}{1+\beta + \beta^2 + \beta^3}   X_t $：

$$ C_t = \frac{1}{1+\beta + \beta^2 + \beta^3}(1+\beta + \beta^2 + \beta^3) \overline{Y}$$

算出：

$$ C_t = \overline{Y}$$

$$(直覺──當利率恆定時，每一期的消費量會等於每一期的恆常所得。)$$

<table><tr><td bgcolor=#98C7DD>----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------</td></tr></table>
