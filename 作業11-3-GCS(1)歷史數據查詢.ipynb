## 匯入GCS 2018-2019(共四季)數據並且重新命名整理

```python
df1=pd.read_csv("GCS.csv",encoding="big5")
df1.head()
```

```python
df1.rename(columns={'Unnamed: 0':'年度','Unnamed: 1':'季度','Unnamed: 2':'隊伍','Unnamed: 3':'選手'}, inplace=True)
```

```python
df1.head()
```

### 查找隊伍歷史數據

用隊伍名字篩選資料後，按照選手名字以及年度排序

```python
team=input("請輸入隊伍名稱：")
dfteam=df1[df1.隊伍 == team]
dfteam.sort_values(by=["選手","年度"],ascending=False)
```
