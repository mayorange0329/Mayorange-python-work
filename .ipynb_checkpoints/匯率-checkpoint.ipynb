# 01查詢匯率

```python
rate = pd.read_html("https://rate.bot.com.tw/xrt/all/day")
```

```python
type(rate)
```

```python
currency = rate[0]
```

```python
#只抓取currency中前五個欄位
currency = currency.iloc[:,:5]
```

```python
currency
```

```python
currency.columns = ['幣別','現金匯率-本行買入','現金匯率-本行賣出','即期匯率-本行賣出','即期匯率-本行買入']
```

```python
currency
```

```python
# 用OrderedDict刪除重複項
from collections import OrderedDict    
currency['幣別'] = [''.join(OrderedDict.fromkeys(v)) for v in currency['幣別']]
```

```python
currency['幣別']
```

```python
currency
```

```python
from ipywidgets import interact_manual,fixed
def f(Currency):
    print(currency[currency.幣別 == Currency])
interact_manual(f,Currency={'美金 (USD)',"港幣 (HKD)",'英鎊 (GBP)','澳幣 (AUD)','加拿大幣 (CAD)','新加坡幣 (SGD)','瑞士法郎 (CHF)',
                            '日圓 (JPY)','南非幣 (ZAR)','瑞典幣 (SEK)','紐元 (NZD)','泰幣 (THB)','菲國比索 (PH)','印尼幣 (IDR)',
                            '歐元 (EUR)','韓元 (KRW)','越南盾 (VND)','馬來幣 (MYR)','人民幣 (CNY)'});
```
