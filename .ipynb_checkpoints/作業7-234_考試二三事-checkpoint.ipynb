```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sympy as sp
from sympy import S
import random
```

### 這是考題很簡單的時候

```python
d1 = pd.DataFrame(columns=['x', 'y'])
d1['x'] = np.random.randint(0, 100, 10)
d1['y'] = np.random.randint(90, 100, 10)
d2 = pd.DataFrame(columns=['x', 'y'])
d2['x'] = np.random.randint(0, 100, 100)
d2['y'] = np.random.randint(70, 100, 100)

plt.scatter(d1['x'], d1['y'], color='red', label='Genius')
plt.scatter(d2['x'], d2['y'], color='grey', label='Mortal')

plt.legend(loc=(1.1, 0))

plt.xlabel("student number")
plt.ylabel("point(s)")

plt.show()
```

### 這是考題很難的時候

```python
d1 = pd.DataFrame(columns=['x', 'y'])
d1['x'] = np.random.randint(0, 100, 10)
d1['y'] = np.random.randint(80, 100, 10)
d2 = pd.DataFrame(columns=['x', 'y'])
d2['x'] = np.random.randint(0, 100, 100)
d2['y'] = np.random.randint(40, 60, 100)

plt.scatter(d1['x'], d1['y'], color='red', label='Genius')
plt.scatter(d2['x'], d2['y'], color='grey', label='Mortal')

plt.legend(loc=(1.1, 0))

plt.xlabel("student number")
plt.ylabel("point(s)")

plt.show()
```

### 我們都曾經以為自己是天才，直到......

```python
 
x=range(1,6)
y=[ [10,20,40,60,80], [60,60,45,30,12], [30,20,15,10,8] ]
 
plt.stackplot(x,y, labels=['Idiot','Mortal','Genius'])
plt.legend(loc=(1.1, 0))
plt.xlabel("As exams get harder and harder→")
plt.ylabel("%")
plt.xticks([])
plt.title("How students think of themselves")

plt.show()
```

### 不過，分數背後總是還有滿滿的故事......

```python
group_names=['High Grade', 'Medium', 'Low Grade']
group_size=[20,60,20]
subgroup_names=['IQ', 'Diligence', 'Luck', 'IQ', 'Study', 'IQ', 'Oversleep', 'Laziness', 'Unlucky' , 'Give up']
subgroup_size=[10,8,2,20,40,8,3,5,1,3]
 
a, b, c=[plt.cm.Greens, plt.cm.Oranges, plt.cm.Blues]
 
# 外面的圈圈
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=[a(0.6), b(0.6), c(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')
 
# 裡面的圈圈
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.7, colors=[a(0.5), a(0.4), a(0.3), b(0.5), b(0.4), c(0.6), c(0.5), c(0.4), c(0.3), c(0.2)])
plt.setp( mypie2, width=0.4, edgecolor='white')

plt.show()
```
