# GCS簡介

[官網](https://esports.moba.garena.tw/gcs)


GCS職業聯賽（Garena Challenger Series）是競舞娛樂《傳說對決》於 2017年7月展開的職業電子競技聯賽。
GCS是台灣第一次以手遊遊戲為主的職業聯賽，每年有春、夏兩季。

```python
import datetime

d1 = datetime.datetime(2017,7,30)   
d2 = datetime.datetime(2019,12,18)  
interval = d2 - d1                  
天數=interval.days                       

infor_GCS=f"截至目前為止，聯賽已經創立{天數}天"

print(infor_GCS)
```

```python
#賽制

from mpl_toolkits.mplot3d import Axes3D

ax = Axes3D(plt.figure())

def f(x, y):
    return (1 - x / 5 + x**2 + y**2) * np.exp(- x**2 - y**2)


x = np.linspace(-2, 4, 100)
y = np.linspace(-2, 4, 100)

X, Y = np.meshgrid(x, y)

plt.contourf(X, Y, f(X, Y), 5, alpha = 0.75, cmap = "rainbow")
C = plt.contour(X, Y, f(X, Y), 5, colors = 'black', linewidth = 0.3)
plt.axis('off')
plt.title("賽制由下至上分別為：業餘賽→城市賽→升降賽→職業賽(GCS)→國際賽(AIC&AWC)",fontproperties="Microsoft JhengHei")
```

GCS賽區在賽事成績上相當有競爭力，拿過兩次AIC國際賽冠軍，與AOG並列最多冠軍的賽區

```python
#各大賽區成績總覽

x=["GCS","AOG","RPL","EU","SM","JP","CN","KR"]
y=[2,2,1,0,0,0,0,1]

plt.stem(x,y)
plt.ylabel("國際賽冠軍次數",fontproperties="Microsoft JhengHei")
plt.xlabel("賽區",fontproperties="Microsoft JhengHei")
```

## GCS數據分析

匯入GCS 2018-2019(共四季)數據並且重新命名整理

```python
df1=pd.read_csv("GCS.csv",encoding="big5")
df1.head()
```

```python
df1.rename(columns={'Unnamed: 0':'年度','Unnamed: 1':'季度','Unnamed: 2':'隊伍','Unnamed: 3':'選手'}, inplace=True)
```

```python
df1.head()
```

### 擊殺數據

```python
總擊殺=df1['擊殺'].sum()
```

```python
平均擊殺=df1['擊殺'].mean()
```

```python
擊殺標準差=df1['擊殺'].std()
```

```python
infor_K=f"截至目前為止，聯賽總共累積了{總擊殺}次擊殺，每個選手平均打出了{平均擊殺}個擊殺數,不過由於各路位置與職責的不同，擊殺數存在著{擊殺標準差}的標準差"

print(infor_K)
```

### 數據分布

我們可以透過箱型圖檢視選手們在擊殺數上的表現分布

```python
fig,axes = plt.subplots()
df1['擊殺'].plot(kind='box',ax=axes, title="Boxplot_Kill")
axes.set_xlabel('Kill')
axes.set_xlabel('Number of kill')
```

### 擊殺王

```python
df1_sorted_K=df1.sort_values(by=["擊殺"],ascending=False)
df1_sorted_K[:1]
```
