# 能力雷達圖

```python
#能力雷達圖
#雷達圖六角名稱
labels = np.array(['發育','打野','承受','團戰','帶線','輸出'])
#雷達圖有六個角
dataLenth = 6
#雷達圖數據
data = np.array([4,3,8,5,4,6])

angles = np.linspace(0, 2*np.pi, dataLenth, endpoint=False)
data = np.concatenate((data, [data[0]]))
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure()
ax = fig.add_subplot(polar=True)
ax.plot(angles, data, 'red', linewidth=2)
ax.fill(angles, data, facecolor='pink')
#雷達圖六角與標題文字，要顯示中文，微軟正黑體
ax.set_thetagrids(angles * 180/np.pi, labels,fontproperties="Microsoft JhengHei")
ax.set_title("能力雷達圖",fontproperties="Microsoft JhengHei")
#雷達圖刻度從零到十
ax.set_rlim(0,10)
ax.grid(True)

plt.show()


HTML('''<script>
code_show=true; 
function code_toggle() {
 if (code_show){
 $('div.input').hide();
 } else {
 $('div.input').show();
 }
 code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
The raw code for this IPython notebook is by default hidden for easier reading.
To toggle on/off the raw code, click <a href="javascript:code_toggle()">here</a>.''')
```
