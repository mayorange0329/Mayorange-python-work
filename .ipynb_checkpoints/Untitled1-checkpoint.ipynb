# 01用股票代號查詢股價

```python
number = input("請輸入股票代號：")
html = pq("https://tw.stock.yahoo.com/q/q?s={}".format(number))
price = html('td[align="center"]').text().split()[3]
print("{}現在的股價是{}".format(html('td[align="center"]').text().split()[0],html('td[align="center"]').text().split()[3]))
```

```python
code=(input("請輸入股票代碼"))
stock=tw.Stock(code)
stock_real=tw.realtime.get(code)
stock_name=stock_real['info']['name']
choose=input("請選擇何種價格？\n1 開盤價\n2 收盤價\n3 最高價\n4 最低價\n")
if choose=='1':
    x='open'
    y='開盤'
elif choose=='2':
    x='latest_trade_price'
    y='收盤'
elif choose=='3':
    x='high'
    y='最高'
elif choose =='4':
    x='low'
    y='最低'
open_price=float(stock_real['realtime'][x])
print(f"今日{stock_name}{y}價格{open_price}")
while True:
    buy=float(input("請輸入當初購買金額"))
    if buy==0:
        break
    earn=((open_price-buy)/buy)*100
    print(f"\033[31m賣出報酬率為{earn:.2f}%\033[0m")
```

```python
df1=pd.read_csv("http://mopsfin.twse.com.tw/opendata/t187ap05_L.csv")
df2=pd.read_csv("http://mopsfin.twse.com.tw/opendata/t187ap05_O.csv")
```

```python
df1.insert(5,"上市櫃","上市")
df2.insert(5,"上市櫃","上櫃")
```

```python
#concat合併
df=pd.concat([df1,df2])
df.drop("出表日期",axis=1,inplace=True)
```

```python
df.sort_values(by="公司代號",inplace=True)
```

```python
df.head()
```

```python
df_sort=df.sort_values(by=["營業收入-去年同月增減(%)"],ascending=[False])
```

```python
def Color(val):
    if val<0:
        color='green' 
    elif val>100:
        color="red"
    else:
        color='black'
    return "color:%s" %color
```

```python
df_sort.head(10).style.applymap(Color,subset=["營業收入-去年同月增減(%)","累計營業收入-前期比較增減(%)","營業收入-上月比較增減(%)"])
```
